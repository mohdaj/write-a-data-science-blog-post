## Cleaning

Before assessing the data to draw conclusions, there are some quality issues that need to be cleaned:
- `Age` colmun has some missing values and in the wrong format
- `LanguageWorkedWith,DevEnviron` must be in the right format
- Standardize the `Gender` column 
- Categorize `Age` to new column 

**1)Fill all the missing values and convert the type to`int` for `Age`**

**Define:**
 `Age` must be in the right fromat for the analysis anf ahs no missing values.

**Code**

```python
df.Age.fillna(df.Age.mean(),inplace=True)
df['Age']=df.Age.astype(int)
```

**Test**

```python
df.Age.sample(3)
```

**2)Convert `LanguageWorkedWith,DevEnviron`to data  typee `str`.**


**Define:**
 `LanguageWorkedWith,DevEnviron` must be in the right fromat for the analysis.

**Code**

```python
#fix languageWorkedwith, DevEnviron datatype, Age_group  
df['LanguageWorkedWith'] = df['LanguageWorkedWith'].astype(str)
df['DevEnviron'] = df['DevEnviron'].astype(str)
```

**Test**

```python
print(df.LanguageWorkedWith.dtype)
print(df.DevEnviron.dtype)
```

**3)Standardize the `Gender` column.**

 **Define:**
     `Gender` has multiple values so it will be grouped into `Man`or`Woman`or`Other`.

**Code**

```python
#fix Gender 
df.Gender.replace('Man;Non-binary, genderqueer, or gender non-conforming','Man',inplace=True)
df.Gender.replace('Woman;Non-binary, genderqueer, or gender non-conforming','Woman',inplace=True)
df.Gender.replace('Non-binary, genderqueer, or gender non-conforming','Other',inplace=True)
df.Gender.replace('Woman;Man','Other',inplace=True)
df.Gender.replace('Woman;Man;Non-binary, genderqueer, or gender non-conforming','Other',inplace=True)
```

**Test**

```python
df.Gender.sample(5)
```

**4)Categorize `Age` to new column.**

 **Define:**
 New colum added to categorizethe `Age` into three Categories`Teen,Adult,Elderly` as `Age_Group`, then converted to the proper format.

**Code**

```python
category = pd.cut(df.Age,bins=[0,20,65,99],labels=['Teen','Adult','Elderly'])
df.insert(5,'Age_Group',category)
df['Age_Group'] = df['Age_Group'].astype(str)
```

**Test**

```python
print(df.Age_Group.value_counts())
print(df.Age_Group.dtype)
```

## Data preparation

```python
def split_column(df, column):
    
    df_copy = df
    series = df_copy[column].apply(lambda x: x.split(';'))
    
    return series

# Splitting the Data Frame by column into a Series. 
lang = split_column(df, 'LanguageWorkedWith')
env = split_column(df,'DevEnviron')
```

```python
def disarray(array_list):
     
    objects = []
    
    for row in array_list:
        for obj in row:
            objects.append(obj.strip())
            
    return objects

# Flatting nested list objects.
lang_list = disarray(lang)
env_list = disarray(env)
```

```python
def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    for x in unique_list: 
        
        
        return unique_list

uni_lang = unique(lang_list)
uni_env = unique(env_list)
```

```python
#removing nan vlues and other
uni_env.remove('nan')
uni_lang.remove('nan')
uni_lang.remove('Other(s):')
```

```python
def assignCount(count,source,factor):
    count[factor]= source.count(factor)
```

```python
langlist={}
envlist={}

for x in uni_lang:
    assignCount(langlist,lang_list,x)

for y in uni_env:
    assignCount(envlist,env_list,y)

print(langlist)
print(envlist)
```

## Exploration

### 1) What are the most common programming language in United States?

```python
lists = sorted(langlist.items(),key=lambda kv: kv[1], reverse=True)
x, y = zip(*lists)
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.suptitle('The most common programming language in United States')
plt.xlabel('Programming language')
plt.ylabel('Count')
plt.xticks(rotation='vertical')
plt.show()
```

**From the graph we can see the top5 common programming languages used in United States are as follows:**
- Java Script
- HTML/CSS
- SQL
- Python 
- Bash/Shell/ PowerShell

## 2)what are the most common development environment in United States?

**From the graph we can see the top5 common Development environment used in United States are as follows:**
